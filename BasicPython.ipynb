```python
import matplotlib.pyplot as plt
import seaborn as sns







```

```python
df1 = pd.read_csv("Universities.csv")
```

```python
plt.hist(df1['SAT'])
plt.xlabel("SAT SCORE")
plt.ylabel("FREQUENCIES")
plt.title("Histogram - ST-FREQUENCY")
```

```python
sns.histplot(df1['SAT'],color='brown')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1['SAT'], df1['Top10'])
plt.xlabel("SAT")
plt.ylabel("Top10")
plt.title("SAT vs. Top10")
plt.show()
```
