```python
print("Program to enter Name and age of a Person and Checks Voting Eligibility")
name = input("Enter Name: ")
age = int(input("Enter Age: "))

if age >= 18:
    print(name, "is eligible to vote")
else:
  print(f"As age of {name} is {age}(< 18), Not eligible to vote")
```

# 6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'


```python
print("Program to Enter marks of a Student and decides Grade")

try:
    marks = float(input("Enter Marks: "))

    if not (0 <= marks <= 100):
        print("Invalid marks! Enter a value between 0 and 100.")
    elif marks >= 90:
        print(f"Marks: {marks} Grade: A")
    elif marks >= 75:
        print(f"Marks: {marks} Grade: B")
    elif marks >= 50:
        print(f"Marks: {marks} Grade: C")
    else:
        print(f"Marks: {marks} Grade: Fail")

except ValueError:
    print("Invalid input! Please enter a numeric value.")
```

#7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
try:
  number = float(input("Enter a number: "))

  if number > 0:
    if number % 2 == 0:
      print(f"{number} is positive and even.")
    else:
      print(f"{number} is positive")
  elif number == 0:
    print(f"The number is {number}")
  else:
    print(f"{number} is negative")

except ValueError:
  print("Invalid input! Please enter a valid number.")
```
