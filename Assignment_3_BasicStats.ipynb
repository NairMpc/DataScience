# Box Plots
Create boxplots for numerical variables to identify outliers and the interquartile range.

Discuss any findings, such as extreme values or unusual distributions.

```python
for column in df_numerical.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df_numerical[column])
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Volume'
Q1 = df_numerical['Volume'].quantile(0.25)
Q3 = df_numerical['Volume'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the bounds for extreme outliers (using 3 * IQR)
# Lower bound for extreme outliers
lower_bound = Q1 - 3 * IQR
# Upper bound for extreme outliers
upper_bound = Q3 + 3 * IQR

# Identify rows with extreme outliers in 'Volume'
extreme_outliers_volume = df_numerical[(df_numerical['Volume'] < lower_bound) | (df_numerical['Volume'] > upper_bound)]

print(f"Q1 for Volume: {Q1}")
print(f"Q3 for Volume: {Q3}")
print(f"IQR for Volume: {IQR}")
print(f"Lower Bound for extreme outliers: {lower_bound}")
print(f"Upper Bound for extreme outliers: {upper_bound}")

print("\nRows with extreme outliers in 'Volume':")
display(extreme_outliers_volume)
```
