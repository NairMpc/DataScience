 #Procedure:

 Mean units produced X_mue = 600

 Sample mean weekly cost W_mean = 3050

 Theoretical mean weekly cost W_mue = 1000 + 5 * 600 = 4000

 Hypothesis.

 restaurant owners' claim that the weekly operating costs are higher than the model suggests.

Ho : W_mue = 4000

Ha : W_mue > 4000

Test statistic:  Z = (W_mean - W_mue)/(sigma/sqrt of n)

given that sigma = 5*25 units, and n=25

And here the sigficance level is given : alpha = 0.05

Being Right tailed test : critical = 1-alpha = 0.95

Z_critical = scipy.stats.norms.ppf(critical)

Then we will compare Z with Z critical

if Z > Z_critical. we will reject Ho and restaurant owners claim(Ha) will be accepted. Else Restaurant owners claim will be rejected.

```python
#Given values
W_mue = 1000 + 5*600
W_mean = 3050
sigma = 5*25
n = 25
alpha = 0.05
```

```python
#Test Statistic
Z = (W_mean - W_mue)/(sigma/(n**0.5))
Z
```

```python
#Find Z_critical
import scipy
Z_critical = scipy.stats.norm.ppf(1-alpha)
Z_critical
```
