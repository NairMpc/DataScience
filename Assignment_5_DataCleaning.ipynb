#● Handle missing values appropriately (e.g., imputation, deletion).
All 14 columns are of integer type. Out of these, 6 columns have no missing values, while 8 columns contain missing values.

Among the columns with missing data, 7 columns have 21 missing values and 1 column has 20 missing values.
Although the number of rows with missing values is relatively small, deleting these rows would result in the loss of valid information from the remaining 6 columns that have no missing values.

Therefore, instead of deletion, imputation is the preferred approach.

The missing values can be filled using appropriate imputation methods after examining the skewness of each column (mean for symmetric distributions and median for skewed distributions).

The absolute value of skewness is used because skewness can be positive or negative, and the decision depends on the magnitude of asymmetry rather than its direction.

```python
SKEW_THRESHOLD = 0.5

for col in df.columns:
  if df[col].isnull().sum() > 0:
    skewness = df[col].skew()

    if abs(skewness) <= SKEW_THRESHOLD:
      fill_value = df[col].mean()
      method = 'Mean'
    else:
      fill_value = df[col].median()
      method = 'Median'

    df[col].fillna(fill_value,inplace=True)
    print(f"{col}:skewness{skewness:.2f}, imputed using {method}")
```

# Observations on Box Plot.


The following code applies capping (winsorization) to handle outliers by replacing values above the upper threshold with the upper bound and values below the lower threshold with the lower bound. This approach reduces the influence of extreme values while preserving the overall structure of the data.

Values greater than Q3+1.5×IQR are capped at the upper bound, and values less than Q1−1.5×IQR are capped at the lower bound. This technique is known as winsorization.

```python
def cap_outliers(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df[columns] = df[columns].apply(lambda x: upper_bound if x > upper_bound else
                                     lower_bound if x < lower_bound else x)

for col in df.columns:
  cap_outliers(df, col)
```

The columns 25-min (which is Q1 - min) and max-75% (which is max - Q3) are equal for many variables because of the outlier capping (winsorization) treatment applied in the data cleaning step.

When we capped outliers, any value below the calculated lower_bound (Q1 - 1.5 * IQR) was replaced by this lower_bound. Similarly, any value above the upper_bound (Q3 + 1.5 * IQR) was replaced by this upper_bound.

After this capping, if a variable originally had outliers that were pushed to these bounds, its new minimum value becomes Q1 - 1.5 * IQR and its new maximum value becomes Q3 + 1.5 * IQR.

Therefore:

25-min becomes Q1 - (Q1 - 1.5 * IQR) = 1.5 * IQR
max-75% becomes (Q3 + 1.5 * IQR) - Q3 = 1.5 * IQR
Since both calculations result in 1.5 * IQR, they become equal. This indicates that for those columns, the data's range has effectively been constrained by the IQR-based outlier boundaries.





The IQR/std ratio can indeed provide some indication of normality, especially for columns where the values are close to the theoretical ratio for a normal distribution.

For a perfectly normal distribution, the IQR is approximately 1.349 times the standard deviation. Therefore, for normally distributed data, the IQR/std ratio should be close to 1.349.

Looking at our summary table:

Columns like LB (1.4098), AC (1.4643), UC (1.5611), DL (1.2327), ASTV (1.6465), MSTV (1.2802), ALTV (1.0598), MLTV (1.2284), Width (1.5920), and Tendency (1.6087) have IQR/std values that are relatively close to 1.349. This suggests that these columns might have distributions that are somewhat similar to a normal distribution, or at least symmetric, especially after the outlier capping applied earlier.

However, columns like FM (1.0329) show a somewhat lower ratio. The values of 0.0 for min, 25%, and 50% for FM, AC, DL, ALTV and Tendency mean that a significant portion of the data is concentrated at zero, which is not characteristic of a normal distribution. Also, DS and DP have NaN for IQR/std because their standard deviation is zero (all values are 0), indicating no variance, which is very far from normal.

It's important to remember that this ratio is just one heuristic. Visual inspections (histograms, Q-Q plots) are also crucial for a more definitive assessment of normality, especially given the outlier capping which can affect the shape of the distribution by compressing the tails.

The correlation heatmap serves as an excellent visual aid to quickly grasp the intensity and direction of correlations between variables.

 The heatmap uses a color scale (e.g., 'coolwarm' in our case) where different hues and shades represent different correlation values:

Warm Colors (e.g., red tones): Typically indicate positive correlations. The brighter or more intense the red, the stronger the positive correlation (closer to +1).

Cool Colors (e.g., blue tones): Typically indicate negative correlations. The darker or more intense the blue, the stronger the negative correlation (closer to -1).

Neutral Colors (e.g., white or light gray): Represent correlations close to zero, indicating a very weak or no linear relationship.

Diagonal: The diagonal cells are usually the most intense color (often bright red if the scale includes 1, or simply distinct) because a variable is perfectly correlated with itself (correlation = 1).

By quickly scanning the heatmap, we can identify highly correlated pairs (bright reds/blues) versus weakly correlated or uncorrelated pairs (neutral colors). For example, a dark blue cell, there's a strong negative relationship, whereas a light red cell suggests a weak positive relationship.
